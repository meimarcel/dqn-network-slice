```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cloud_first = np.array(
    [
        [0, 0, 0],
        [9830, 0, 0],
        [17694, 0, 0],
        [23592, 0, 0],
        [25558, 0, 0],
        [29490, 0, 0],
        [29490, 0, 0],
        [37354, 0, 0],
        [39994.4, 1291.6, 0],
    ]
)
cloud_first_lost_package = [0, 0, 1, 3, 7, 10, 15, 16, 19]
cloud_first_energy = [0, 600, 600, 600, 600, 600, 600, 600, 900]

fog_first = np.array(
    [
        [0, 0, 0],
        [370, 9460, 0],
        [740, 18920, 0],
        [1110, 18920, 9460],
        [1480, 18920, 18920],
        [5412.0, 18920.0, 18920],
        [5412.0, 18920.0, 18920],
        [13276.0, 18920.0, 18920],
        [23106.0, 18920.0, 18920],
    ]
)
fog_first_lost_package = [0, 0, 0, 0, 0, 3, 8, 9, 9]
fog_first_energy = [0, 900, 900, 1200, 1200, 1200, 1200, 1200, 1200]

dqn = np.array(
    [
        [0, 0, 0],
        [9830, 0, 0],
        [17768, 1892, 0],
        [22522.4, 6967.6, 0],
        [22892.4, 16427.6, 0],
        [25154.4, 18319.6, 5676],
        [25524.4, 18319.6, 15136],
        [33462.4, 18319.6, 17028],
        [39417.600000000006, 19611.199999999997, 19611.199999999997],
    ]
)
dqn_lost_package = [0, 0, 0, 0, 0, 0, 0, 0, 0]
dqn_energy = [0, 600, 900, 900, 900, 1200, 1200, 1200, 1200]

time_steps = np.arange(len(dqn))
```

```python
# Cloud band
bar_width = 0.25
r1 = np.arange(len(time_steps))
r2 = [x + bar_width for x in r1]
r3 = [x + 2 * bar_width for x in r1]

plt.figure(figsize=(12, 6))
plt.bar(r1, cloud_first[:, 0], width=bar_width, label="Cloud Priority", color="blue", alpha=0.7)
plt.bar(r2, fog_first[:, 0], width=bar_width, label="Fog Priority", color="red", alpha=0.7)
plt.bar(r3, dqn[:, 0], width=bar_width, label="DQN", color="green", alpha=0.7)

plt.xlabel("Time Step", fontweight="bold", fontsize=20)
plt.ylabel("Bands (Mbps)", fontweight="bold", fontsize=20)
plt.title("Cloud Band Over Time", fontweight="bold", fontsize=20)
plt.xticks([r + bar_width for r in range(len(time_steps))], time_steps, fontsize=20)
plt.yticks(fontsize=20)
plt.legend(prop={"weight": "bold", "size": 14})
plt.savefig("cloud_band_over_time.eps", dpi=400)
plt.show()
```

```python
# Fog band
bar_width = 0.25
r1 = np.arange(len(time_steps))
r2 = [x + bar_width for x in r1]
r3 = [x + 2 * bar_width for x in r1]

plt.figure(figsize=(12, 6))
plt.bar(r1, cloud_first[:, 1:].sum(1), width=bar_width, label="Cloud Priority", color="blue", alpha=0.7)
plt.bar(r2, fog_first[:, 1:].sum(1), width=bar_width, label="Fog Priority", color="red", alpha=0.7)
plt.bar(r3, dqn[:, 1:].sum(1), width=bar_width, label="DQN", color="green", alpha=0.7)

plt.xlabel("Time Step", fontweight="bold", fontsize=20)
plt.ylabel("Bands (Mbps)", fontweight="bold", fontsize=20)
plt.title("Fog Band Over Time", fontweight="bold", fontsize=20)
plt.xticks([r + bar_width for r in range(len(time_steps))], time_steps, fontsize=20)
plt.yticks(fontsize=20)
plt.legend(prop={"weight": "bold", "size": 14})
plt.savefig("fog_band_over_time.eps", dpi=400)
plt.show()
```
